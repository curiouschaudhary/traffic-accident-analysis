```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Set a style for all plots

```python
sns.set_style('whitegrid')
```

# Data Loading and Initial Inspection

```python
df = pd.read_csv('traffic accident.csv')
df
```

```python
# Display the first 5 rows of the dataframe
print("--- First 5 Rows ---")
print(df.head())
```

```python
# Get a concise summary of the dataframe, including data types and non-null values
print("\n--- DataFrame Info ---")
df.info()
```

```python
# Get descriptive statistics for numerical columns
print("\n--- Descriptive Statistics ---")
print(df.describe())
```

# Data Cleaning and Preprocessing

```python
# Convert the 'crash_date' column to datetime objects
df['crash_date'] = pd.to_datetime(df['crash_date'])
```

# --- Feature Engineering ---

```python
# Extract new features from the 'crash_date' column
df['crash_year'] = df['crash_date'].dt.year
df['crash_month'] = df['crash_date'].dt.month
df['crash_day_of_week'] = df['crash_date'].dt.dayofweek  # Monday=0, Sunday=6
df['crash_hour'] = df['crash_date'].dt.hour
```

#  Exploratory Data Analysis (EDA)

```python
# --- Univariate Analysis ---

# Distribution of Weather Conditions
plt.figure(figsize=(12, 7))
sns.countplot(y='weather_condition', data=df, order=df['weather_condition'].value_counts().index)
plt.title('Distribution of Accidents by Weather Condition', fontsize=16)
plt.xlabel('Number of Accidents', fontsize=12)
plt.ylabel('Weather Condition', fontsize=12)
plt.show()
```

```python
# Distribution of Lighting Conditions
plt.figure(figsize=(12, 7))
sns.countplot(y='lighting_condition', data=df, order=df['lighting_condition'].value_counts().index)
plt.title('Distribution of Accidents by Lighting Condition', fontsize=16)
plt.xlabel('Number of Accidents', fontsize=12)
plt.ylabel('Lighting Condition', fontsize=12)
plt.show()
```

```python
# Distribution of Accidents by Hour of the Day
plt.figure(figsize=(12, 6))
sns.histplot(df['crash_hour'], bins=24, kde=True)
plt.title('Distribution of Accidents by Hour of the Day', fontsize=16)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(range(0, 24))
plt.grid(axis='y', alpha=0.5)
plt.show()
```

# --- Bivariate Analysis ---

```python
# Heatmap of Accidents by Day of Week and Hour
day_hour_crosstab = pd.crosstab(df['crash_day_of_week'], df['crash_hour'])
plt.figure(figsize=(14, 8))
sns.heatmap(day_hour_crosstab, cmap='viridis', annot=False)
plt.title('Heatmap of Accidents by Day of Week and Hour', fontsize=16)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Day of the Week (Mon=0, Sun=6)', fontsize=12)
plt.show()
```

# --- Multivariate Analysis ---

```python
# Select only the numerical columns for the correlation matrix
numerical_cols = df.select_dtypes(include=np.number)
plt.figure(figsize=(12, 10))
sns.heatmap(numerical_cols.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features', fontsize=16)
plt.show()
```
